**<font color=black> Adenda, Aquí unas lineas para leer un archivo plano utilizando la función open y for para leer cada lines del archivo. Por inquietud del estudiante Andrés Camilo incluímos unas lineas para borrar unas lineas y guardar el archivo sin esas líneas en un nuevo archivo** 

```python
f=open('C:/Users/usuario/Documents/in.comp.txt', 'r') # Abrimos el archivo
f1=f.readlines() # Leemos cada linea del archivo y se guarda en f1 
i=0 # Usamos un contador para saber la posición hasta donde queremos borrar líneas
target="# ------------------------ INITIALIZATION ----------------------------"# Usamos un identificador para saber en donde empieza
                                                                               #el nuevo archivo
for line in f1:
    if target in line: # Leemos las líneas 
        break # y detenemos la secuencia cuando encontramos el identificador
    else:
        i=i+1 # Contamos las líneas
with open('C:/Users/usuario/Documents/in.comp.txt') as old, open('C:/Users/usuario/Documents/n-in.comp.txt', 'w') as new:
    lines = old.readlines() # Ahora lleemos todas las líneas
    new.writelines(lines[i:-1]) # Y generamos un nuevo archivo sin tener encuenta las lineas iniciales hasta el identificador
```

```python
data_bar2= pd.read_excel('C:/Users/Enrique Mejia/Documents/Data_processing/crudo3/V_all_molecules.xlsx', usecols='B,C, D')
data_bar2.head(10)
```

```python
dff = (data_bar2[data_bar2['DBE'].eq(10)])
dff
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
colors = plt.cm.tab10(range(20)) #Dark2, Pastel1, Paired, Accent, Pastel2, Set1, Set2, Set3, tab10, tab20, tab10b
def draw_barchart(dbe):
    dff = data_bar2[data_bar2['DBE'].eq(dbe)].sort_values(by='Rel. Abudance', ascending=True).tail(10)
    ax.clear()
    ax.barh(dff['MF'], dff['Rel. Abudance'], color=colors)
    dx = dff['Rel. Abudance'].max() / 200
    for i, (value, name) in enumerate(zip(dff['Rel. Abudance'], dff['MF'])):
        ax.text(value-dx, i, name, size=14, weight=600, ha='right', va='bottom')
        #ax.text(value-dx, i-.25, name, size=10, color='#444444', ha='right', va='baseline')
        ax.text(value+dx, i, f'{value:,.0f}',  size=14, ha='left',  va='center')
    # ... polished styles
    ax.text(1, 0.4, dbe, transform=ax.transAxes, color='#777777', size=46, ha='right', weight=800)
    ax.text(0, 1.06, 'Rel. Abundance (arb. units)', transform=ax.transAxes, size=12, color='#777777')
    #ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'DBE FT-ICR-MS Vasconia crude oil from 10 to 30',
            transform=ax.transAxes, size=24, weight=600, ha='left')
    plt.box(True)
    
draw_barchart(10)
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
fig, ax= plt.subplots(figsize=(12, 7))
animator = animation.FuncAnimation(fig, draw_barchart, frames=range(5, 20))
HTML(animator.to_jshtml()) 
```
