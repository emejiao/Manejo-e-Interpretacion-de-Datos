```python
data1_HC['HC']= data1_HC['Unnamed: 8'] + data1_HC['Unnamed: 10']# Crear una columna, hay que asegurarse que son del mismo
data1_HC.head() #                                                 tipo
```

```python
# Es posible convertir las columnas entre tipos de datos, en este ejemplo, en string
data1_HC[['Unnamed: 8','Unnamed: 10']] = data1_HC[['Unnamed: 8','Unnamed: 10']].astype(str)
```

```python
data1_HC['F. Molecular']= data1_HC['Molecular Formula'] + data1_HC['Unnamed: 8'] + data1_HC['Unnamed: 9'] + data1_HC['Unnamed: 10']# Crear una columna, hay que asegurarse que son del mismo
data1_HC.head() #                                                 tipo
```

```python
# Borramos columnas
data1_HC=data1_HC.drop(['Molecular Formula', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'HC', 'Unnamed: 0'], axis=1)
data1_HC.head()
```

<font color='red' size=20>**Datos perdidos**

```python
# Vamos a leer un archivo que contiene datos perdidos
flights = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/flights.csv")
flights.head()
```

```python
# Podemos inspeccionar todas las filas que contienen al menos un NaN usando la función isnull()
flights[flights.isnull().any(axis=1)].head()
```

```python
# Podemos remover todas las observaciones que contienen NaN usando la función dropna
flights2 = flights.dropna()
```

```python
# O podemos llenar los valores perdidos con ceros o cualquier otro valor
nomiss =flights.fillna(0)
nomiss.isnull().any()
```

<font color=red size=4>**Otras formas...**

```python
# Elimina todo los elementos perdidos desde las filas que los contienen:
flights.dropna(axis=1, how='all')
```

```python
# Elimina todos los elementos perdidos desde las filas y columnas
flights.dropna(axis=1, how='any')
```

```python
# Solo se mantienen las filas que contengan como máximo dos datos perdidos
flights.dropna(thresh=2)
```

```python
# Llena los valores perdidos con el valor promedio de cada columna
flights.fillna(flights.mean())
```

```python
# Una Columna específica 
flights['dep_time'].fillna(flights['dep_time'].median())
```
