<font color='green' size=6>**Filtración y Selección**

```python
data1_HC= data1['HC']
data1_HC.head()
```

```python
DBE_M10= data1_HC[data1_HC["DBE"]>10] # Vamos a filtrar utilizando la columna DBE y valor de ésta mayor a 10
DBE_M10.head()
```

```python
DBE_m10= data1_HC[data1_HC["DBE"]<10] # Vamos a filtrar utilizando la columna DBE y valor de ésta menor a 10
DBE_m10.head()
```

```python
DBE_I10= data1_HC[data1_HC["DBE"]==10] # Vamos a filtrar utilizando la columna DBE y valor de ésta igual a 10
DBE_I10.head()
```

```python
# Filtrado de columnas y filas 
DBE_HC= data1_HC[["DBE","H/C"]][1:25]
DBE_HC
```

```python
data1_HC['HC']= data1_HC['Unnamed: 8'] + data1_HC['Unnamed: 10']# Crear una columna, hay que asegurarse que son del mismo
data1_HC.head() #                                                 tipo
```

```python
# Es posible convertir las columnas entre tipos de datos, en este ejemplo, en string
data1_HC[['Unnamed: 8','Unnamed: 10']] = data1_HC[['Unnamed: 8','Unnamed: 10']].astype(str)
```

```python
data1_HC['F. Molecular']= data1_HC['Molecular Formula'] + data1_HC['Unnamed: 8'] + data1_HC['Unnamed: 9'] + data1_HC['Unnamed: 10']# Crear una columna, hay que asegurarse que son del mismo
data1_HC.head() #                                                 tipo
```

```python
# Borramos columnas
data1_HC=data1_HC.drop(['Molecular Formula', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'HC', 'Unnamed: 0'], axis=1)
data1_HC.head()
```

---
<font color='red' size=6>**Ordenamiento**

```python
# Los datos pueden ser ordenados con respecto a los valores de una de las columnas
data1_HC_sorted = data1_HC.sort_values(by = 'DBE')
data1_HC_sorted.head(20)
```

```python
# Ordenamiento de los datos y sobreescritura de los datos original
data1_HC.sort_values(by = 'DBE', ascending = False, inplace = True)
data1_HC.head()
```

```python
# restauración del orden original de los datos
data1_HC.sort_index(axis=0, ascending = True, inplace = True)
data1_HC.head()
```

```python
# Ordenamiento usando 2 o mas columnas:
data1_HC_sorted2 = data1_HC.sort_values(by = ['DBE', 'H/C'], ascending = [True,False])
data1_HC_sorted2.head(10)
```
